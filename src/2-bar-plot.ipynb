### Exercise 2: Bar Plot with Colors

```python
import matplotlib.pyplot as plt

# Define the categories and values
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'blue', 'green', 'purple']

# Check if categories, values, and colors arrays have the same length
if not (len(categories) == len(values) == len(colors)):
    raise ValueError("categories, values, and colors arrays must have the same length")

# Create the bar plot with different colors for each bar
plt.bar(categories, values, color=colors)

# Customize the plot
plt.title('Bar Plot of Categories and Values')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.grid(True)

# Display the plot
plt.show()
```
