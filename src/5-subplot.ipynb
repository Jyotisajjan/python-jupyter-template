### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the plots
x = np.arange(1, 6)
y = x * 2
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
colors = ['red', 'blue', 'green', 'purple']  # Define the colors array

# Check if arrays used for the plots have the correct lengths
if len(x) != len(y):
    raise ValueError("x and y arrays must have the same length")
if not (len(categories) == len(values) == len(colors)):
    raise ValueError("categories, values, and colors arrays must have the same length")
if len(labels) != len(sizes):
    raise ValueError("labels and sizes arrays must have the same length")

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Line chart in the first subplot
axs[0, 0].plot(x, y)
axs[0, 0].set_title('Line Chart')
axs[0, 0].set_xlabel('x values')
axs[0, 0].set_ylabel('y values')

# Bar chart in the second subplot
axs[0, 1].bar(categories, values, color=colors)
axs[0, 1].set_title('Bar Chart')
axs[0, 1].set_xlabel('Categories')
axs[0, 1].set_ylabel('Values')

# Scatter plot in the third subplot
axs[1, 0].scatter(x, y)
for i in range(len(x)):
    axs[1, 0].annotate(f'({x[i]}, {y[i]})', (x[i], y[i]), textcoords="offset points", xytext=(0,10), ha='center')
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].set_xlabel('x values')
axs[1, 0].set_ylabel('y values')

# Pie chart in the fourth subplot
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
axs[1, 1].set_title('Pie Chart')

# Adjust layout
plt.tight_layout()

# Display the plot
plt.show()
```
